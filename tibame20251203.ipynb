```python
#自然對數的底數 (\(e\)): 約等於 \(2.71828\)，是自然指數函數的底數。
#e=2.71828182845904523536，近似值為271801/99990{\frac {271801}{99990}}}。
# 1/1 +1/1 +1/1*2+ 1/1*2*3 +1/1*2*3*4
total = 0
fraction = 1
i = 0
while i < 10000:
    if not i == 0:
        fraction = fraction * i
    total = total + 1 / fraction
    #print(total)
    i = i + 1
print(total)
```

```python
import ColabTurtle.Turtle as turtle
#盡量不要寫from 來import，因為會怕蓋過其他

turtle.initializeTurtle()
# 存起來： x, y = turtle.pos()
# goto：turtle.goto(x, y)

turtle.speed(8)
init_width = 60
shape_count = 8
shape_angle = 360 / shape_count
points = [0] * shape_count
i = 0
j = 0

#turtle.pencolor("brown")
#從中心出發畫出支線
while i < shape_count:
    # 往上畫直線(存起來)
    if not i ==0:
        turtle.right(shape_angle)
    turtle.forward(init_width)
    points[i] = turtle.pos()
    #print(points[i])
    if i < shape_count -1:
        turtle.backward(init_width)
    i = i + 1

#turtle.bgcolor("#efefef")
turtle.pencolor('red')
#把存下來的點依序連起外框
while j < shape_count:
    p_x, p_y = points[j]
    turtle.goto(p_x, p_y)
    j = j + 1


```
