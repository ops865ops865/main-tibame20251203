```python
import ColabTurtle.Turtle as turtle

turtle.initializeTurtle()
turtle.speed(5)
init_width = 80
shape_count = 10 #設定多邊形的形狀
shape_angle = 360 / shape_count #設定每個三角形中心的角度
points = [0] * shape_count #存取多邊形的點
turtle.pencolor("brown")

#只用一個回圈完成
startx, starty = turtle.pos()
i = 0
while i < shape_count:
    turtle.forward(init_width)
    firstx, firsty = turtle.pos()
    turtle.goto(startx, starty)
    turtle.right(shape_angle)
    turtle.forward(init_width)
    turtle.goto(firstx, firsty)
    turtle.goto(startx, starty)
    #因為goto是平行移動，所以角度不用是對的
    i += 1
```

```python
import ColabTurtle.Turtle as turtle

# 座標系統 左上角(0, 0) 往右x增加 往下y增加
turtle.initializeTurtle(initial_window_size=(400,400))
turtle.speed(5)

# 第一個大正方形，四個點的list
points = [
    [50, 50],
    [350, 50],
    [350, 350],
    [50, 350]
]

k = 0
while k < 5: #隨便設定，畫五次
    turtle.penup()
    x, y = points[0]
    turtle.goto(x, y)
    turtle.pendown()
    # 寫一個四次的迴圈
    i = 0
    print(len(points))
    while i < len(points):
        x, y = points[(i+1) % len(points)]
        turtle.goto(x, y)
        print(x, y)
        i = i + 1

    # 畫完正方形你才可以做取半
    # 創建一個新的list  [-1] * 4
    # 算出新的半點  list[0] = 座標
    half_points = [-1] * 4
    j = 0
    while j < len(points):
        x, y = points[j]
        next_x, next_y =points[(j+1)%len(points)]
        half_x, half_y = (x + next_x) /2, (y + next_y) /2
        half_points[j] = [half_x, half_y]
        #turtle.goto(half_x, half_y)
        j = j + 1
    # 算出新的四個 points = 新的list
    points = half_points  #！！！我忘記這個了
    k = k + 1
```

```python
#https://www.unixtimestamp.com/
#timestamp = 1764000000
#20251124 16:00:00
timestamp = 1764894708
#20251205 00:31:48

start_year = 1970
# seconds = 365 * 24 * 60 * 60 #31536000
# more_seconds = 366 * 24 * 60 * 60 #31622400
SECONDS_PER_MINUTE = 60
SECONDS_PER_HOUR = 60 * SECONDS_PER_MINUTE
SECONDS_PER_DAY = 24 * SECONDS_PER_HOUR

#timestamp = 63072001
#先判斷是否為閏年(366天)，然後扣掉當年的秒
while True:
    if is_leap_year(start_year):
        year_day = 366
    else:
        year_day = 365
    year_second = year_day * SECONDS_PER_DAY
    if timestamp / year_second >= 1:
        timestamp = timestamp - year_second
        start_year +=1
    else:
        break
# while timestamp / seconds >= 1:
#     if is_leap_year(start_year):
#         timestamp = timestamp - 366 * SECONDS_PER_DAY
#     else:
#         timestamp = timestamp - 365 * SECONDS_PER_DAY
#     start_year +=1

#目前可算出年份
print(start_year)
print(timestamp)

#要計算出月份
days_per_month =[31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
month = 1
#閏年2月要加一天
if is_leap_year(start_year):
    days_per_month[1] = 29
while month <= 12:
    if timestamp < SECONDS_PER_DAY * days_per_month[month-1]:
        break
    else:
        timestamp = timestamp - SECONDS_PER_DAY * days_per_month[month-1]
        month += 1
#目前可算出月份
print(month)
print(timestamp)

#要計算出日期，取商為日期
day = (timestamp // SECONDS_PER_DAY) +1
timestamp = timestamp % SECONDS_PER_DAY
print(day)
print(timestamp)

#要計算出小時，取商為小時
hour = timestamp // SECONDS_PER_HOUR
timestamp = timestamp % SECONDS_PER_HOUR
print(hour)
print(timestamp)

#要計算出分鐘、秒，取商為分鐘，取餘為秒數
minute = timestamp // SECONDS_PER_MINUTE
timestamp = timestamp % SECONDS_PER_MINUTE
second = timestamp
print(minute)
print(timestamp)

print("%d-%d-%d,%d:%d:%d" % (start_year,month,day,hour,minute,second))
print("{}-{}-{},{}:{}:{}".format(start_year,month,day,hour,minute,second))
print(str(start_year) + '-' + str(month) + '-' + str(day) +','+ str(hour) + ':'+ str(minute) + ':' + str(second))
```
