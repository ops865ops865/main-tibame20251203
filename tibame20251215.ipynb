```python
#[
#   [0, 1, 2, 3],
#   [7, 6, 5, 4],
#   [8, 9, 10, 11]
#]
# ans[0][0]=0
# ans[0][1]=1
# ans[0][2]=2
# ans[0][3]=3
# ans[1][3]=4
# ans[1][2]=5

w, h = 4, 6
ans = [[0] * w for i in range(h)]
total = 0

for j in range(h):
    for i in range(w):
        if j%2 ==0:
            ans[j][i]=total
        else:
            ans[j][abs(i-3)]=total
        total+=1
ans
# ------------------------------------
w, h = 4, 8
ans = [[0] * w for i in range(h)]

for i in range(h):
    for j in range(w):
        if i % 2 == 0:
            ans[i][j] = w * i + j
        else:
            ans[i][j] = w * i + (w - 1) - j
ans
```
